```python
print('Название фильма                     | Год  | Рейтинг | Длина | Бюджет | Сборы  |')
print('--------------------------------------------------------------------------------')
for row in films[:]:
    print('{: <35} | {} | {: >7.2f} | {: >5} | {: >6.1f} | {: >6.1f} |'.format(
        row[0], row[1], row[2], row[3], row[5], row[6]))
```

## Сортировка данных

```python
films.sort(key=lambda i: i[1], reverse=True)

print('Название фильма                     | Год  | Рейтинг | Длина | Бюджет | Сборы  |')
print('--------------------------------------------------------------------------------')
for row in films[:]:
    print('{: <35} | {} | {: >7.2f} | {: >5} | {: >6.1f} | {: >6.1f} |'.format(
        row[0], row[1], row[2], row[3], row[5], row[6]))
```

## 1. Задача

```python
start_year = 1900
end_year = 2020

def column_sum(films, column, start_year, end_year):
    result = 0
    for row in films:
        if start_year <= row[1] <= end_year:
            result += row[column]
    return result
column = 5                        # 5 - бюджет, 6 - кассовый сбор
print(column_sum(films, column, start_year, end_year))
```
